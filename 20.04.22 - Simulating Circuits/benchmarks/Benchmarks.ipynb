# Benchmark Data

```python
import qiskit
import qiskit.quantum_info as qi
import numpy as np

def ghz_circuit(num_qubits):
    """Return an N-qubit GHZ preparation circuit."""
    circ = qiskit.QuantumCircuit(num_qubits)
    circ.h(0)
    for j in range(1, num_qubits):
        circ.cx(0, j)
    return circ
```

## Circuit Simulation

### Clifford class

```python
nqs_cliff = list(range(1, 26))
ts_cliff = []

print('Clifford')
for nq in nqs_cliff:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.Clifford(circ)
    ts_cliff.append(t.average)

print('\n Times (s)')
print(ts_cliff)
```

### Statevector class

```python
nqs_sv = list(range(1, 26))
ts_sv = []

print('Statevector')
for nq in nqs_sv:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.Statevector.from_instruction(circ)
    ts_sv.append(t.average)

print('\n Times (s)')
print(ts_sv)
```

### DensityMatrix class

```python
nqs_dm = list(range(1, 14))
ts_dm = []

print('Density Matrix')
for nq in nqs_dm:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.DensityMatrix.from_instruction(circ)
    ts_dm.append(t.average)

print('\n Times (s)')
print(ts_dm)
```

### Operator class

```python
nqs_op = list(range(1, 15))
ts_op = []

print('Operator')
for nq in nqs_op:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.Operator(circ)
    ts_op.append(t.average)

print('\n Times (s)')
print(ts_op)
```

### SuperOp class

```python
nqs_sop = list(range(1, 8))
ts_sop = []

print('SuperOp')
for nq in nqs_sop:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.SuperOp(circ)
    ts_sop.append(t.average)

print('\n Times (s)')
print(ts_sop)
```

## Experiment Simulation

### Aer QasmSimulator (stabilizer)

```python
# AER QASM STABILIZER
nqs_counts_aer_stab = list(range(1, 31))
ts_counts_aer_stab = []

print('Aer QasmSimulator("stabilizer")')
for nq in nqs_counts_aer_stab:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    circ.measure_all()
    opts = {"method": "stabilizer"}
    sim = qiskit.Aer.get_backend('qasm_simulator')
    t = %timeit -o -r 1 qiskit.execute(circ, sim, shots=1000, backend_options=opts).result().get_counts(0)
    ts_counts_aer_stab.append(t.average)

print('\n Times (s)')
print(ts_counts_aer_stab)
```

### Aer QasmSimulator (statevector)

```python
nqs_counts_aer_sv = list(range(1, 31))
ts_counts_aer_sv = []

print('Aer QasmSimulator("statevector")')
for nq in nqs_counts_aer_sv:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    circ.measure_all()
    opts = {"method": "statevector"}
    sim = qiskit.Aer.get_backend('qasm_simulator')
    t = %timeit -o -r 1 qiskit.execute(circ, sim, shots=1000, backend_options=opts).result().get_counts(0)
    ts_counts_aer_sv.append(t.average)

print('\n Times (s)')
print(ts_counts_aer_sv)
```

### Statevector class

```python
nqs_counts_sv = list(range(1, 21))
ts_counts_sv = []

print('Statevector')
for nq in nqs_counts_sv:
    print('{}-qubits'.format(nq))
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.Statevector.from_instruction(circ).sample_counts(1000)
    ts_counts_sv.append(t.average)

print('\n Times (s)')
print(ts_counts_sv)
```

## Clifford large N Timing

```python
nqs_large_cliff = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
ts_cliff_circ = []
ts_cliff_sim = []

for nq in nqs_large_cliff:
    print('{}-qubits'.format(nq))

    # Time circuit
    t = %timeit -o -r 1 ghz_circuit(nq)
    ts_cliff_circ.append(t.average)
    
    # Time Clifford sim
    circ = ghz_circuit(nq)
    t = %timeit -o -r 1 qi.Clifford(circ)
    ts_cliff_sim.append(t.average)

# Save data
print('\n Time (s) Cliford circuit')
print(ts_cliff_circ)
print('\n Time (s) Cliford simulation')
print(ts_cliff_sim)
```
